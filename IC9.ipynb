## 1. Extracting key phrases for each review.

**Keyword extraction with TextRank**

```python
from summa import keywords
```

```python
yelp['keyphrase_1'] = yelp['text'].apply(lambda x: keywords.keywords(x).split("\n"))
yelp.head()
```

**Keywords Extraction with YAKE!**

```python
language = "en"
max_ngram_size = 3
deduplication_threshold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 5

from yake import KeywordExtractor
kw_extractor = KeywordExtractor(lan=language, n=max_ngram_size, top=numOfKeywords,dedupLim=deduplication_threshold,dedupFunc=deduplication_algo, windowsSize=windowSize,features=None)
```

```python
yelp['keyphrase_2'] = yelp['text'].apply(lambda x: kw_extractor.extract_keywords(x))
yelp.head()
```

## 2. Use Spacy to extract Person entities for each review.

```python
import spacy as sp
nlps = sp.load("en_core_web_sm")
```

```python
yelp['Spacy_person'] = yelp['text'].apply(lambda x: [tok for tok in nlps(x).ents if tok.label_ == 'PERSON'])
```

```python
yelp.head()
```

## 3. Use Spacy to extract ORG entities for each review.

```python
yelp['Spacy_org'] = yelp['text'].apply(lambda x: [tok for tok in nlps(x).ents if tok.label_ == 'ORG'])
yelp.head()
```

## 4. Use NLTK to extract terms with NNP tagging for each review.

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
yelp['NLTK_NNP'] = yelp['text'].apply(lambda x: [word for word, pos in pos_tag(word_tokenize(x)) if pos == 'NNP'])
yelp.head()
```

## 5. Use Spacy to extract relationships for each review.

```python
def extract_relationships(doc):
    # Find all the subject-verb-object triples in the doc
    subject_verb_pairs = []
    for possible_subject in doc:
        if possible_subject.dep_ == "nsubj":
            verb = possible_subject.head
            if verb.dep_ == "ROOT":
                subject_verb_pairs.append((possible_subject, verb))

    relationships = []
    for subject, verb in subject_verb_pairs:
        for possible_object in verb.children:
            if possible_object.dep_ == "dobj":
                relationships.append((subject.text, verb.lemma_, possible_object.text))

    return relationships
```

```python
yelp['Spacy_RE'] = yelp['text'].apply(lambda x: extract_relationships(nlps(x)))
yelp.head()
```

## 6. Use Duckling to extract numbers for each review.

```python
from duckling import DucklingWrapper
```

```python
d = DucklingWrapper()
```

```python
yelp['Duckling_number'] = yelp['text'].apply(lambda x: [item['text'] for item in d.parse_number(x)])
yelp.head()
```

## 7. Use Duckling to extract currency for each review.

```python
yelp['Duckling_currency'] = yelp['text'].apply(lambda x: [item['text'] for item in d.parse_money(x)])
yelp.head()
```
