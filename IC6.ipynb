(5.2) Using Naive Bayes to predict the sentiment score of the first 1000 rows of reviews, given that `sentiment_score` = 1 if `overall` > 3.

```python
df.overall.unique()
```

```python
df['sentiment_score'] = df['overall'].apply(lambda x: 1 if x>3 else 0)
df
```

```python
df2 = df.head(1000)
```

```python
import numpy as np
import nltk
import string
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
wn = WordNetLemmatizer()

def clean_text(text):
    words = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split("\W+", words)
    text = [wn.lemmatize(word) for word in tokens if word not in stopwords]
    return text
```

```python
countvec = CountVectorizer(analyzer = clean_text)
X_count = countvec.fit_transform(df2["Review_Text"])

df3 = pd.DataFrame(X_count.toarray(), columns=countvec.get_feature_names_out())
df3['label'] = df2['sentiment_score']
df3
```

```python
features = df3.drop(['label'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(features, df3['label'], test_size = 0.2, random_state = 1)
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train, y_train)
y_pred_class = naive_bayes.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred_class))
```

Therefore, the Naive Bayes has an accuracy of 0.87.
